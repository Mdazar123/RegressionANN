```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
import pickle
```

```python
data=pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data
```

```python
label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
data
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo=OneHotEncoder()
geo_encoder=onehot_encoder_geo.fit_transform(data[['Geography']])
```

```python
geo_encoder.toarray()
```

```python
geo_encoder_df=pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoder_df
```

```python
# data=pd.concat([data.drop('Geography',axis=1),geo_encoder_df],axis=1)
# data.head()
data
```

```python
x=data.drop('EstimatedSalary',axis=1)
y=data['EstimatedSalary']
# splittig the features into training and testing sets
x = pd.get_dummies(x, drop_first=True)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
# scale this feautres
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
with open('scaler_reg.pkl','wb') as file:
  pickle.dump(scaler, file)
```

```python
# train the model with Regression neural network
import tensorflow as tf #type:ignore
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model=Sequential([
  Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
  Dense(32,activation='relu'),
  Dense(1)
        #  no activationg for regression default linear activation
])

model.compile(optimizer="adam",loss='mean_absolute_error',metrics=['mae'])

model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
log_dir='regressionlogs/fit/'+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callbacks=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=10,restore_best_weights=True)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Define EarlyStopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Optionally, define ModelCheckpoint callback
model_checkpoint = ModelCheckpoint('my_model.keras', monitor='val_loss', save_best_only=True)
```

```python
history=model.fit(
  x_train,y_train,validation_data=(x_test,y_test),epochs=100,
  callbacks=[tensorflow_callbacks,early_stopping]
)
```

```python
# eval model on test data
test_loss,test_mae=model.evaluate(x_test,y_test)
print(f"test mae :{test_mae}")
```
